```python
import time
```

```python
import random
```

```python
# 모든 데이터에 대해 가져오기
total_data_list = []
page_num = 1
prev_data_time_set = set()

while True :
    url = "https://finance.naver.com/item/sise_day.naver?code=005930&page={}".format(page_num)
    res = requests.get(url, headers = req_headers)
    soup = bs4.BeautifulSoup(res.text)
    data_tr_elements = soup.select("table.type2 > tr[onmouseover='mouseOver(this)']")
    
    # 각 page 의 모든 정보를 담는 list
    page_data_list = []
    
    # 현재 페이지 날짜 정보만 담는 set
    current_data_time_set = set()
    
    # 각 행을 순회한다.
    for e in data_tr_elements :
        
        # 각 tr, 즉 row 마다 data 를 가져온다
        td_elements = e.select("td")
        
        # 행에 있는 모든 td 데이터들을 집어넣는다.
        data_list = []
        
        # enumerate : idx~val 을 쌍으로 리턴해준다.
        for idx, td_e in enumerate(td_elements) :
            input_td_e = td_e.text.strip()    
            # 공백 X
            if input_td_e == "":
                continue
            # 날짜 데이터
            if idx == 0 :
                current_data_time_set.add(td_e)
                data_list.append(input_td_e)
            # 숫자 형태로 데이터를 추가하기 , 중간에 ',' 제거하기
            else :
                data_list.append(int(input_td_e.replace(",", "")))
        
        # 데이터가 존재할 경우에만 append
        if len(data_list) > 0 :
            page_data_list.append(data_list)

    # 범위를 넘어서는 page 를 crawling 하면, 이전 page 와 같은 내용을 가져온다.
    if prev_data_time_set == current_data_time_set :
        print("끝")
        break;
        
    prev_data_time_set = current_data_time_set
    
    # total_data_list 을 2차원 list 형태로 담기 위함
    for data in page_data_list :
        total_data_list.append(data)
        
    # 지나치게 빠른 속도로 요청을 보내는 것을 제한하기 위함
    time.sleep(random.random() * 2)
    page_num += 1
    if page_num == 3 :
        break;
```

```python
# 내가 가져온 데이터의 날짜가, 가장 마지막으로 저장한 날짜와 동일하면 중단
# 마지막 row 추출
# df.iloc[-1]
```

```python
from datetime import datetime
```

```python
# 현재는 날짜가 문자열 형태로 인식된다.
# 날짜 형태로 인식하기 위해서 변형할 것이다.
# "%Y.%m.%d" 는 얼마든지 변형해도 된다. 그저 현재 우리가 크롤링한
# 데이터가 2023.04.25 형태여서 이렇게 진행한 것이다.
last_date_time = datetime.strptime(opened_df.iloc[-1]['날짜'], "%Y.%m.%d")
```

```python
last_date_time
```

```python
# 모든 데이터에 대해 가져오기
new_total_data_list = []
page_num = 1
prev_data_time_set = set()

while True :
    url = "https://finance.naver.com/item/sise_day.naver?code=005930&page={}".format(page_num)
    res = requests.get(url, headers = req_headers)
    soup = bs4.BeautifulSoup(res.text)
    data_tr_elements = soup.select("table.type2 > tr[onmouseover='mouseOver(this)']")
    
    # 각 page 의 모든 정보를 담는 list
    page_data_list = []
    
    # 현재 페이지 날짜 정보만 담는 set
    current_data_time_set = set()
    
    # 각 행을 순회한다.
    for e in data_tr_elements :
        
        # 각 tr, 즉 row 마다 data 를 가져온다
        td_elements = e.select("td")
        
        # 행에 있는 모든 td 데이터들을 집어넣는다.
        data_list = []
        
        # enumerate : idx~val 을 쌍으로 리턴해준다.
        for idx, td_e in enumerate(td_elements) :
            input_td_e = td_e.text.strip()    
            # 공백 X
            if input_td_e == "":
                continue
            # 날짜 데이터
            if idx == 0 :
                # 날짜 데이터를 문자열이 아닌, 실제 날짜 포멧으로 넣어준다.
                input_td_e = datetime.strptime(input_td_e, "%Y.%m.%d")
                current_data_time_set.add(input_td_e)
                data_list.append(input_td_e)
            # 숫자 형태로 데이터를 추가하기 , 중간에 ',' 제거하기
            else :
                data_list.append(int(input_td_e.replace(",", "")))
        
        # 데이터가 존재할 경우에만 append
        if len(data_list) > 0 :
            page_data_list.append(data_list)

    # 범위를 넘어서는 page 를 crawling 하면, 이전 page 와 같은 내용을 가져온다.
    if prev_data_time_set == current_data_time_set :
        print("끝")
        break;
        
    prev_data_time_set = current_data_time_set
    
    last_date_included = False
    # 현재 페이지에서 긁어온 데이터 날짜 중에서
    # 내가 가장 마지막에 긁어온 데이터가 존재한다면
    # 그 이후의 겻만 total_data_list 에 넣기 
    if last_date_time in current_data_time_set :
        last_date_included = True
        
    # total_data_list 을 2차원 list 형태로 담기 위함
    for data in page_data_list :
        if last_date_included and data[0] <= last_date_time :
            continue
        new_total_data_list.append(data)
    
    if last_date_included :
        break;
        
    # 지나치게 빠른 속도로 요청을 보내는 것을 제한하기 위함
    time.sleep(random.random() * 2)
    page_num += 1
    if page_num == 3 :
        break;
```

```python
new_df = pd.DataFrame(new_total_data_list,columns= opened_df.columns)
```
